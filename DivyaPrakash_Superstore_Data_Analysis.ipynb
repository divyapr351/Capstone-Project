```python
## Import pandas as pd
import pandas as pd
```

```python
## Load the dataset
sp_df = pd.read_csv ("/content/SampleSuperstore.csv")
```

Data Visualization

```python
sp_df.head(10)
```

```python
sp_df.describe()
```

```python
## Sales analysis by region
sp_df.groupby("Region")["Sales"].sum().plot.line()
```

```python
## Sales analysis by State
sp_df.groupby("State")["Sales"].sum().plot.bar()
```

```python
### Profit analysis by Region
sp_df.groupby("Region")["Profit"].sum().plot.bar()
```

```python
## Profit analysis by State
sp_df.groupby("State")["Profit"].sum().plot.bar()
```

```python
## Sales analysis by Category
sp_df.groupby("Category")["Sales"].sum().plot.bar()
```

```python
## Profit analysis by Category
sp_df.groupby("Category")["Profit"].sum().plot.bar()
```

```python
## Sales analysis by sub-category
sp_df.groupby("Sub-Category")["Sales"].sum().plot.bar()
```

```python
## Profit analysis by sub-category
sp_df.groupby("Sub-Category")["Profit"].sum().plot.bar()
```
